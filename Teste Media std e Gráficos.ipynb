```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler os dados do arquivo CSV
data = pd.read_csv('media_std.csv')

# Criar uma figura e um eixo
plt.figure(figsize=(12, 8))

# Configurações do gráfico
bar_width = 0.6  # Largura das barras
error_kw = {'linewidth': 2, 'capsize': 5}  # Configuração para o erro padrão

# Plotar gráfico de barras para 'acc'
bars = plt.barh(data['Model'], data['acc'], xerr=data['acc_std'], color='skyblue', edgecolor='black', height=bar_width, error_kw=error_kw)

# Adicionar valores na barra
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.02, bar.get_y() + bar.get_height() / 2, f'{width:.2f}', va='center', ha='left', fontsize=10, color='black')

# Configurar rótulos e título
plt.xlabel('Accuracy (acc)', fontsize=14)
plt.ylabel('Model', fontsize=14)
plt.title('Accuracy (acc) com Desvio Padrão dos Modelos', fontsize=16)
plt.tight_layout()

# Salvar o gráfico
plt.savefig('grafico_acc.png', dpi=300)
plt.show()
```
