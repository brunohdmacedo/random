```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Configurações gerais
num_particulas = 100
num_frames = 100
tempo_simulacao = 10  # segundos
dt = tempo_simulacao / num_frames

# Função para gerar partículas
def gerar_particulas():
    posicoes = np.random.rand(num_particulas, 2) * 10  # posições aleatórias
    velocidades = np.random.randn(num_particulas, 2)  # velocidades aleatórias
    return posicoes, velocidades

# Função para atualizar partículas no sistema canônico
def atualizar_particulas_canonico(posicoes, velocidades):
    posicoes += velocidades * dt
    # Simulação simples de colisão com a cunha (reflexão)
    for i, (x, y) in enumerate(posicoes):
        if x < 0 or y < 0:  # Colisão com as bordas (cunha triangular simples)
            velocidades[i] = -velocidades[i]  # Reflexão simples
    return posicoes, velocidades

# Função para atualizar partículas no sistema microcanônico
def atualizar_particulas_microcanonico(posicoes, velocidades):
    posicoes += velocidades * dt
    # Simulação simples de colisão com a cunha (reflexão)
    for i, (x, y) in enumerate(posicoes):
        if x < 0 or y < 0:  # Colisão com as bordas (cunha triangular simples)
            velocidades[i] = -velocidades[i]  # Reflexão simples
    return posicoes, velocidades

# Função de animação
def animar(frame, posicoes_canonico, velocidades_canonico, posicoes_microcanonico, velocidades_microcanonico):
    plt.clf()

    # Atualiza as partículas
    posicoes_canonico, velocidades_canonico = atualizar_particulas_canonico(posicoes_canonico, velocidades_canonico)
    posicoes_microcanonico, velocidades_microcanonico = atualizar_particulas_microcanonico(posicoes_microcanonico, velocidades_microcanonico)

    # Plotagem das partículas
    plt.scatter(posicoes_canonico[:, 0], posicoes_canonico[:, 1], c='b', label='Canônico')
    plt.scatter(posicoes_microcanonico[:, 0], posicoes_microcanonico[:, 1], c='r', label='Microcanônico')

    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Simulação de Partículas')
    plt.xlim(0, 10)
    plt.ylim(0, 10)
    plt.legend()
    plt.grid(True)

    return plt.gca().lines

# Inicializa a posição e velocidade das partículas
posicoes_canonico, velocidades_canonico = gerar_particulas()
posicoes_microcanonico, velocidades_microcanonico = gerar_particulas()

# Cria a animação
fig = plt.figure(figsize=(8, 6))
ani = animation.FuncAnimation(fig, animar, frames=num_frames, fargs=(posicoes_canonico, velocidades_canonico, posicoes_microcanonico, velocidades_microcanonico), interval=100, repeat=False)
ani.save('simulacao_particulas.gif', writer='pillow')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# Parâmetros da simulação
num_particles = 500  # Número de partículas
m = 1.0  # Massa das partículas
e = 1.0  # Coeficiente de restituição (colisão elástica)
box_size = 10.0  # Tamanho da caixa
particle_radius = 0.1  # Raio das partículas

# Inicializar posições e velocidades aleatórias
np.random.seed(0)
positions = np.random.rand(num_particles, 2) * box_size - box_size / 2
velocities = (np.random.rand(num_particles, 2) - 0.5) * 2

# Criar a figura e os eixos
fig, ax = plt.subplots()
ax.set_xlim(-box_size / 2, box_size / 2)
ax.set_ylim(-box_size / 2, box_size / 2)
particles = ax.scatter(positions[:, 0], positions[:, 1], c='b', s=5)

# Lista para armazenar a energia cinética total ao longo do tempo
total_kinetic_energy_list = []
velocities_list = []  # Lista para armazenar as velocidades

def init():
    particles.set_offsets(np.empty((0, 2)))
    particles.set_array(np.array([]))
    return particles,

def update(frame):
    dt = 0.02
    global positions, velocities

    # Atualizar posições
    positions += velocities * dt

    # Detectar colisões com as paredes
    for i in range(num_particles):
        if positions[i, 0] <= -box_size / 2 or positions[i, 0] >= box_size / 2:
            velocities[i, 0] = -velocities[i, 0]
        if positions[i, 1] <= -box_size / 2 or positions[i, 1] >= box_size / 2:
            velocities[i, 1] = -velocities[i, 1]

    # Detectar colisões entre partículas
    for i in range(num_particles):
        for j in range(i + 1, num_particles):
            if np.linalg.norm(positions[i] - positions[j]) < 2 * particle_radius:
                # Calcular o vetor unitário ao longo da linha de impacto
                k = (positions[j] - positions[i]) / np.linalg.norm(positions[j] - positions[i])
                # Calcular a velocidade relativa inicial
                relative_velocity_initial = velocities[i] - velocities[j]
                # Calcular a componente da velocidade relativa na direção do impacto
                relative_velocity_normal = np.dot(relative_velocity_initial, k) * k
                # Calcular a velocidade relativa tangencial (não afetada pela colisão)
                relative_velocity_tangential = relative_velocity_initial - relative_velocity_normal
                # Calcular a velocidade normal final após a colisão
                relative_velocity_normal_final = -e * relative_velocity_normal
                # Calcular a velocidade final das partículas
                velocities[i] += (relative_velocity_normal_final - relative_velocity_normal) / 2
                velocities[j] -= (relative_velocity_normal_final - relative_velocity_normal) / 2

    # Calcular energia cinética total e atualizar cores
    kinetic_energy = 0.5 * m * np.sum(velocities**2, axis=1)
    total_kinetic_energy_list.append(np.sum(kinetic_energy))  # Armazenar a energia cinética total
    velocities_list.append(velocities.copy())  # Armazenar as velocidades
    colors = plt.cm.viridis(kinetic_energy / np.max(kinetic_energy))

    particles.set_offsets(positions)
    particles.set_color(colors)
    return particles,

# Criar a animação
anim = FuncAnimation(fig, update, frames=300, init_func=init, blit=True)

# Salvar a animação como GIF
gif_path = 'particles_collide_microcanonical.gif'
anim.save(gif_path, writer=PillowWriter(fps=20))

plt.close(fig)

# Plotar a energia cinética total ao longo do tempo
plt.figure()
plt.plot(total_kinetic_energy_list)
plt.xlabel('Frame')
plt.ylabel('Energia Cinética Total')
plt.title('Energia Cinética Total ao Longo do Tempo (Microcanônico)')
plt.grid(True)
plt.savefig('energia_cinetica_total_microcanonico.png')
plt.show()

# Coletar todas as velocidades para a distribuição
velocities_flat_microcanonical = np.linalg.norm(np.vstack(velocities_list), axis=1)

# Plotar a distribuição das velocidades
plt.figure()
plt.hist(velocities_flat_microcanonical, bins=30, density=True, alpha=0.6, color='g')
plt.xlabel('Velocidade')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição das Velocidades (Microcanônico)')
plt.grid(True)
plt.savefig('distribuicao_velocidades_microcanonico.png')
plt.show()

gif_path
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# Parâmetros da simulação
num_particles = 100
m = 1.0  # massa das partículas
e = 0.8  # coeficiente de restituição (colisão parcialmente inelástica)
dt = 0.01  # passo de tempo

# Definir condições iniciais
positions = np.random.rand(num_particles, 2) * [10, 10] - [5, 5]
velocities = np.random.randn(num_particles, 2) * 0.5

# Definir geometria da cunha
wedge_x = np.array([0, 5, 10])
wedge_y = np.array([0, 5, 0])

# Funções para detectar colisões com a cunha
def detect_wedge_collision(pos):
    if pos[0] < 0 or pos[0] > 10:
        return False
    if pos[0] < 5:
        return pos[1] < pos[0]
    else:
        return pos[1] < 10 - pos[0]

def reflect_wedge(pos, vel):
    if pos[0] < 5:
        normal = np.array([1, -1]) / np.sqrt(2)
    else:
        normal = np.array([-1, -1]) / np.sqrt(2)
    vel_normal = np.dot(vel, normal) * normal
    vel_tangential = vel - vel_normal
    vel_reflected = -e * vel_normal + vel_tangential
    return vel_reflected

# Criar a figura e os eixos
fig, ax = plt.subplots()
ax.set_xlim(-5, 15)
ax.set_ylim(-5, 10)
particles, = ax.plot([], [], 'bo', ms=2)
ax.plot(wedge_x, wedge_y, 'k-')

def init():
    particles.set_data([], [])
    return particles,

def update(frame):
    global positions, velocities

    # Atualizar posições
    positions += velocities * dt

    # Detectar e tratar colisões com a cunha
    for i in range(num_particles):
        if detect_wedge_collision(positions[i]):
            velocities[i] = reflect_wedge(positions[i], velocities[i])

    # Detectar e tratar colisões entre partículas
    for i in range(num_particles):
        for j in range(i + 1, num_particles):
            if np.linalg.norm(positions[i] - positions[j]) < 0.1:
                k = (positions[j] - positions[i]) / np.linalg.norm(positions[j] - positions[i])
                relative_velocity = velocities[i] - velocities[j]
                relative_velocity_normal = np.dot(relative_velocity, k) * k
                relative_velocity_tangential = relative_velocity - relative_velocity_normal
                relative_velocity_normal_final = -e * relative_velocity_normal
                velocities[i] += (relative_velocity_normal_final - relative_velocity_normal) / 2
                velocities[j] -= (relative_velocity_normal_final - relative_velocity_normal) / 2

    particles.set_data(positions[:, 0], positions[:, 1])
    return particles,

# Criar a animação
anim = FuncAnimation(fig, update, frames=1000, init_func=init, blit=True)

# Salvar a animação como GIF
gif_path = 'particulas_colidem_com_cunha.gif'
anim.save(gif_path, writer=PillowWriter(fps=30))

plt.close(fig)

gif_path
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir parâmetros da simulação
num_particles = 100  # Número de partículas
time_steps = 1000    # Número de passos de tempo
dt = 0.01            # Intervalo de tempo

# Inicializar posições e velocidades das partículas
positions = np.random.rand(num_particles, 2) * 10  # Posições aleatórias
velocities = np.random.randn(num_particles, 2)  # Velocidades aleatórias

# Definir a geometria da asa do avião (simplesmente uma linha horizontal)
wing_y = 5.0  # Altura da asa

# Função para detectar e tratar colisões com a asa do avião
def handle_collisions(positions, velocities):
    for i in range(num_particles):
        if positions[i, 1] <= wing_y:  # Checar colisão com a asa
            velocities[i, 1] = -velocities[i, 1]  # Inverter a velocidade na direção y

# Registrar velocidades para plotagem
velocities_history = np.zeros((time_steps, num_particles, 2))

# Loop de simulação
for t in range(time_steps):
    positions += velocities * dt  # Atualizar posições
    handle_collisions(positions, velocities)  # Tratar colisões
    velocities_history[t] = velocities  # Registrar velocidades

# Plotar gráficos das velocidades
plt.figure(figsize=(12, 6))
for i in range(num_particles):
    plt.plot(velocities_history[:, i, 1], label=f'Particle {i}')
plt.xlabel('Time step')
plt.ylabel('Velocity')
plt.title('Velocities of Particles over Time')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```
